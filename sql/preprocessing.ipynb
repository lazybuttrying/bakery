```python
import datetime as dt
import random

sql = []
userId = {0:'user01',1:'ewha777'}
seasonal = {12:'winter',1:'winter',2:'winter',
         3:'spring',4:'spring',5:'spring',
         6:'summer',7:'summer',8:'summer',
         9:'fall',10:'fall',11:'fall'}
day_of_week = [0,"mon","tue","wed","thu","fri","sat","sun"]

def change_date(x):
    global seasonal
    date = list(map(int, x.split()[0].split('-')))
    s = seasonal[date[1]]
    date = dt.datetime(date[0], date[1], date[2])
    d = day_of_week[date.isoweekday()]
    y = date.year

    time = list(map(int, x.split()[1].split(':')))
    timezone = ''
    if time[0]<=11:
        timezone = 'M'
    elif time[0]>=16:
        timezone = 'E'
    else:
        timezone = "N"
        
    return s, d, timezone, str(y)

def location(x):
    if x=="동내면":
        return random.choice(["신촌리","거두리"])
    elif x=="동면":
        return random.choice(["만천리","장학리","지내리"])
    elif x=="신동면":
        return random.choice(["증리","정족리"])
    return x
```
